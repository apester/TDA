# Lecture 1 — Simplicial Complexes

```python
import numpy as np
import matplotlib.pyplot as plt

triangle = np.array([[0,0],[1,0],[0.5,0.8]])

plt.fill(triangle[:,0],triangle[:,1],alpha=0.3)
plt.plot(*np.vstack([triangle,triangle[0]]).T,marker='o')
plt.title("2-simplex example")
plt.axis("equal")
plt.show()
```
