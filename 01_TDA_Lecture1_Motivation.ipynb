# Lecture 1 — Motivation for Topological Data Analysis
This notebook demonstrates why **shape of data** matters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_circles
from sklearn.cluster import KMeans

np.random.seed(42)
plt.rcParams["figure.figsize"] = (6,6)
```

## Blob dataset

```python
X_blobs,_ = make_blobs(n_samples=300, centers=[(-2,-2),(2,2),(-2,2)], cluster_std=0.6)
plt.scatter(X_blobs[:,0], X_blobs[:,1], s=15)
plt.title("Three clusters")
plt.axis("equal")
plt.show()
```

## Circle dataset

```python
X_circle,_ = make_circles(n_samples=400, factor=0.75, noise=0.05)
plt.scatter(X_circle[:,0], X_circle[:,1], s=15)
plt.title("Circle point cloud")
plt.axis("equal")
plt.show()
```

## k-means on circle

```python
kmeans = KMeans(n_clusters=2, n_init=10)
labels = kmeans.fit_predict(X_circle)

plt.scatter(X_circle[:,0], X_circle[:,1], c=labels, s=15)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker="x", s=200)
plt.title("k-means on circle")
plt.axis("equal")
plt.show()
```
